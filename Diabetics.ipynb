```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

it is possible to determine the dimensions of a dataset by looking at the shape of a pandas DataFrame, the functio returns a tuple consisting of two elements 
(Tuple is an immutable and ordered collection of elemens in python, used to store multiple items in a single variable, simillar to lists):
-Dataset rows: the number of observations of patients
-The number of columns: indicsates the amount of features or attriubutes each patient has

What are the benefits of shape?
-A dataset's SIZE indicates how big it is
-COMPLEXITY: the number of features that need to be analyzed or cleaned

What is the value if information?
    - **Overview of the dataset**: It immediately tells us how many entries we are analysing, which is crucial for understanding its scope.
    - **Memory and performance**: Knowing the shape helps us estimate the processing resources required.
    - **Validation of data**: Ensures datasets have the expected number of observations and features (for example, if a dataset should have 10 features but has 15, it indicates a problem).
    - **Feature complexity**: The number of features that require analysis or cleaning.
